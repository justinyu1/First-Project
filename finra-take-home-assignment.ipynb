# Circuit Breaker Requirement 10

```python
# Circuit Breaker Requirement no. 10
mask1a = df1['trd_actn_cd'] == '0'
mask1b = df1['trd_actn_cd'] == '1'
mask1c = df1['trd_actn_cd'] == '2'
mask2 = df1['msg_ctgry_cd'] == 'A'
mask3 = df1['msg_type_cd'] == 'V'
```

```python
utp_a = df1.loc[mask1a & mask2 & mask3]
# utp_a meets the conditions below
# trd_actn_cd = 0, msg_ctgry_cd = A, msg_type_cd = V
```

```python
utp_b = df1.loc[mask1b & mask2 & mask3]
# utp_b meets the conditions below
# trd_actn_cd = 1, msg_ctgry_cd = A, msg_type_cd = V
```

```python
utp_c = df1.loc[mask1c & mask2 & mask3]
# utp_c meets the conditions below
# trd_actn_cd = 2, msg_ctgry_cd = A, msg_type_cd = V
```

```python
utp_issue_reqs = pd.concat([utp_a, utp_b, utp_c])
utp_issue_reqs
# utp_issue_reqs = new variable circuit breaker requirement 10
# concatenated 3 dataframes with the varying trd_actn_cd values
```

# Circuit Breaker Requirement 20 conditions 1 & 2

```python
# Merged utp_issues_reqs and df2 where a.issue_sym_id = b.issue_sym_id
df_merged = pd.DataFrame.merge(utp_issue_reqs, df2, on = ['issue_sym_id'])
df_merged
```
